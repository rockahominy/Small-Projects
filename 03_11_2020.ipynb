```python
from scipy.stats import norm
import matplotlib.pyplot as pt
```

```python
import pandas as p
iris = p.read_csv('path.csv')

set_sp = iris.loc[iris['species'] == 'setosa']
vir_sp = iris.loc[iris['species'] =='virginica']
ver_sp = iris.loc[iris['species'] == 'versicolor']
```

```python
#sep_len_set = set_sp['sepal_length']
```

```python
import matplotlib.pyplot as pt

fig, ax = pt.subplots()
blue = ax.plot(set_sp['sepal_length'], marker = 'o', color = 'b', label = 'setosa')
red = ax.plot(ver_sp['sepal_length'], marker = 'o', color = 'r', label = 'versicolor')
green = ax.plot(vir_sp['sepal_length'], marker = 'o', color = 'g', label = 'virginica')
ax.set_xlabel("Observations for Irisâ€”Setosa, Virginica, and Versicolor")
ax.set_ylabel("Sepal Length")
ax.set_title("Sepal Length for all Iris species")
ax.legend([blue, red, green], frameon= True, loc = 'best')
pt.show()
```
